```python
import sys
sys.path.append('../') # go to parent dir
from collections import OrderedDict
%matplotlib inline
from matplotlib import pyplot as plt
import torchvision
import torch
import torch.nn as nn
from networks.style_gan_net import MappingNet, SynthesisNet
```

```python
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
generator = nn.Sequential(OrderedDict([
    ('g_mapping', MappingNet()),
    ('g_synthesis', SynthesisNet())
]))
generator.load_state_dict(torch.load('../pretrained/karras2019stylegan-ffhq-1024x1024.for_g_all.pt'))
generator.to(device)
generator.eval()
```

```python
torch.manual_seed(7)
latents = torch.randn(2, 512, device=device)
```

```python
with torch.no_grad():
    imgs = generator(latents)
    imgs = (imgs.clamp(-1, 1) + 1) / 2.0

imgs = imgs.cpu()
imgs = torchvision.utils.make_grid(imgs, nrow=1)

plt.figure(figsize=(15, 6))
plt.imshow(imgs.permute(1, 2, 0).detach().numpy())
```
